```python
df_cluster = pd.DataFrame()
```

```python
df_cluster['Credit amount'] = data['Credit amount']
df_cluster['Age'] = data['Age']
df_cluster['Duration'] = data['Duration']
df_cluster['Job'] = data['Job']

# onlay which data we want to copay that much we copy....
```

```python
df_cluster_log = np.log(df_cluster[['Credit amount','Age','Duration']])  #
# we using numpy is here ..
```

```python
from sklearn.preprocessing import StandardScaler  #data (cluster) ko scale ker rhai hai

scaler = StandardScaler()
cluster_scaled = scaler.fit_transform(df_cluster_log)
```

```python
from sklearn.cluster import KMeans
```

```python
Sum_of_squared_distances =[]

K =  range(1,15)
for k in K:
    km=KMeans(n_clusters=k)
    km=km.fit(cluster_scaled)
    Sum_of_squared_distances.append(km.inertia_)
    
    
#    see the pdf EDA  - k mean = cluster ma mean sa point ka closer jata hai 
#     best possible means of number..  \best optimesd number
```

```python
X1.shape
```

```python
plt.figure(figsize=(20,5))
plt.plot(K,Sum_of_squared_distances,'bx-')

# elabo methode here we use one more methode is knee method that see...
# best optimesd number is 2 ....
```

```python
# lebaling for the grph

plt.figure(figsize=(20,5))
plt.plot(K,Sum_of_squared_distances,'bx-')
plt.ylabel('Sum_of_squared_distances')
plt.title("Elbow Methode for Optimal K Value")
plt.show()
```
